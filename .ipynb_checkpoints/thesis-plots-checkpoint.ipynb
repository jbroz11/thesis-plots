```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.art3d as art3d
from matplotlib.patches import Circle, PathPatch
from matplotlib.text import TextPath
from matplotlib.transforms import Affine2D
from mpl_toolkits.mplot3d import axes3d
from matplotlib import rc

from scipy.special import factorial
import numpy as np
```

```python
# https://matplotlib.org/3.3.0/gallery/mplot3d/pathpatch3d.html
def text3d(ax, xyz, s, zdir="z", size=None, angle=0, usetex=False, **kwargs):
    '''
    Plots the string 's' on the axes 'ax', with position 'xyz', size 'size',
    and rotation angle 'angle'.  'zdir' gives the axis which is to be treated
    as the third dimension.  usetex is a boolean indicating whether the string
    should be interpreted as latex or not.  Any additional keyword arguments
    are passed on to transform_path.

    Note: zdir affects the interpretation of xyz.
    '''
    x, y, z = xyz
    if zdir == "y":
        xy1, z1 = (x, z), y
    elif zdir == "x":
        xy1, z1 = (y, z), x
    else:
        xy1, z1 = (x, y), z

    text_path = TextPath((0, 0), s, size=size, usetex=usetex)
    trans = Affine2D().rotate(angle).translate(xy1[0], xy1[1])

    p1 = PathPatch(trans.transform_path(text_path), **kwargs)
    ax.add_patch(p1)
    art3d.pathpatch_2d_to_3d(p1, z=z1, zdir=zdir)
```

```python
AC = True

fig = plt.figure(figsize=plt.figaspect(1)*3)
ax = fig.add_subplot(111, projection="3d", computed_zorder=False)
X, Y = np.mgrid[-1/2:1/2:50j, -1/2:1/2:50j]
X1, Y1 = np.mgrid[-0.9:0.9:50j, -0.9:0.9:50j]
if AC:
    Z = 1.5 + X**2 + Y**2
    Z1 = 1 + X1**2 + Y1**2
else:
    Z1 = 1.5 + X1**2 - Y1**2
    Z = 2 - X**2 + Y**2

# surface 
stride = 5
alpha = 1 if AC else 0.9
ax.plot_surface(X1, Y1, Z1, lw=0.4, edgecolor="k", rstride=stride, cstride=stride, alpha=alpha, color="gray", zorder=1)

#contour
ax.contour(X, Y, Z, 7, cmap="gray", linestyles="solid", offset=0, alpha=1, corner_mask=False)

# axes
axes_extent = 2.4
x = np.linspace(0, axes_extent, 1000)
zero = 0*x
ax.plot(x, zero, color="blue", zorder=-1)
ax.plot(-x, zero, color="blue", zorder=-1)
ax.plot(zero, x, color="blue", zorder=-1)
ax.plot(zero, -x, color="blue", zorder=-1)

# axes arrows
asize = 0.2
a1 = plt.arrow(0, axes_extent, 0, 0.1, shape='full', lw=0, length_includes_head=True, head_width=asize, head_length=asize, color="blue")
art3d.pathpatch_2d_to_3d(a1, z=0, zdir="z")
a1 = plt.arrow(axes_extent, 0, 0.1, 0, shape='full', lw=0, length_includes_head=True, head_width=asize, head_length=asize, color="blue")
art3d.pathpatch_2d_to_3d(a1, z=0, zdir="z")

# axes labels
text3d(ax, (axes_extent + 0.1, 0, 0),
           r"$\hat{x}_1$",
           zdir="y", size=0.5, usetex=True,
           ec="none", fc="blue", angle=0)
text3d(ax, (0, axes_extent + 0.1, 0),
           r"$\hat{x}_2$",
           zdir="y", size=0.5, usetex=True,
           ec="none", fc="blue", angle=0)

# electrodes
ccolor = "k"
p1 = plt.Polygon([(-1, 0), (-1.5, -0.25), (-1.5, 0.25)], facecolor=ccolor, lw=2, edgecolor="k")
ax.add_patch(p1)
art3d.pathpatch_2d_to_3d(p1, z=0, zdir="z")
s1 = plt.Rectangle((-1.5, -0.25), -0.5, 0.5, facecolor=ccolor, lw=2, edgecolor="k")
ax.add_patch(s1)
art3d.pathpatch_2d_to_3d(s1, z=0, zdir="z")
p2 = plt.Polygon([(1, 0), (1.5, -0.25), (1.5, 0.25)], facecolor=ccolor, lw=2, edgecolor="k")
ax.add_patch(p2)
art3d.pathpatch_2d_to_3d(p2, z=0, zdir="z")
s2 = plt.Rectangle((1.5, 0.25), 0.5, -0.5, facecolor=ccolor, lw=2, edgecolor="k")
ax.add_patch(s2)
art3d.pathpatch_2d_to_3d(s2, z=0, zdir="z")
p3 = plt.Polygon([(0, 1), (-0.25, 1.5), (0.25, 1.5)], facecolor=ccolor, lw=2, edgecolor="k")
ax.add_patch(p3)
art3d.pathpatch_2d_to_3d(p3, z=0, zdir="z")
s3 = plt.Rectangle((0.25, 1.5), -0.5, 0.5, facecolor=ccolor, lw=2, edgecolor="k")
ax.add_patch(s3)
art3d.pathpatch_2d_to_3d(s3, z=0, zdir="z")
p4 = plt.Polygon([(0, -1), (-0.25, -1.5), (0.25, -1.5)], facecolor=ccolor, lw=2, edgecolor="k")
ax.add_patch(p4)
art3d.pathpatch_2d_to_3d(p4, z=0, zdir="z")
s4 = plt.Rectangle((0.25, -1.5), -0.5, -0.5, facecolor=ccolor, lw=2, edgecolor="k")
ax.add_patch(s4)
art3d.pathpatch_2d_to_3d(s4, z=0, zdir="z")

# electrode connections
ax.plot([-1.75, -1.75], [0, -2.5], [0, 0], color="k")
ax.plot([1.75, 1.75], [0, -2.5], [0, 0], color="k")
ax.plot([-1.75, 1.75], [-2.5, -2.5], [0, 0], color="k")
ax.plot([0, 0], [-1.75, -4], [0, 0], color="k")
ax.plot([0, 3], [1.75, 1.75], [0, 0], color="k")
ax.plot([0, 3], [-4, -4], [0, 0], color="k")
ax.plot([3, 3], [1.75, -4], [0, 0], color="k")
ax.plot([3, -1.75], [-4, -4], [0, 0], color="k")
c1 = plt.Circle((-1.75, -2.5), 0.1, color="k")
ax.add_patch(c1)
art3d.pathpatch_2d_to_3d(c1, z=0, zdir="z")
c2 = plt.Circle((0, -4), 0.1, color="k")
ax.add_patch(c2)
art3d.pathpatch_2d_to_3d(c2, z=0, zdir="z")

# ground
ax.plot([0, 0], [-4, -4], [0, -1/2], color="k")
g = plt.arrow(0, -1/2, 0, -0.1, shape='full', lw=0, length_includes_head=True, head_width=asize, head_length=asize, color="k")
art3d.pathpatch_2d_to_3d(g, z=-4, zdir="y")
text3d(ax, (0.2, -4, -0.6),
       "ground",
       zdir="y", size=0.4, usetex=True,
       ec="none", fc="k", angle=0)

if AC:
    # AC signal
    ax.plot([-1.75, -1.75], [-2.5, -2.75], [0, 0], color="k")
    ax.plot([-1.75, -1.75], [-3.75, -4], [0, 0], color="k")
    ac = plt.Circle((-1.75, -3.25), 0.5, edgecolor="k", facecolor="white", lw=2)
    ax.add_patch(ac)
    art3d.pathpatch_2d_to_3d(ac, z=0, zdir="z")
    actheta = np.linspace(0, 2 * np.pi, 1000)
    acx = np.linspace(-1.5, -2, 1000)
    ax.plot(acx, -0.2 * np.sin(actheta) - 3.25, zero*acx, color="k", lw=3)
    text3d(ax, (-3.9, -3.25, 0),
           r"$V$" + "cos" + r"$(\Omega t)$",
           zdir="y", size=0.4, usetex=True,
           ec="none", fc="k", angle=0)
    # # ion
    # rs = 0.1
    # us = np.linspace(0, 2 * np.pi, 100)
    # vs = np.linspace(0, np.pi, 100)
    # xs = rs * np.outer(np.cos(us), np.sin(vs))
    # ys = rs * np.outer(np.sin(us), np.sin(vs))
    # zs = rs * np.outer(np.ones(np.size(us)), np.cos(vs))
    # ax.plot_surface(xs, ys, zs + 2.1, color='red', alpha=1, zorder=-1)
    
else:
    # DC signal
    width = 0.3
    ax.plot([-1.75, -1.75], [-2.5, -3.25 + width/2], [0, 0], color="k")
    ax.plot([-1.75, -1.75], [-3.25 - width/2, -4], [0, 0], color="k")
    ax.plot([-2.25, -1.25], [-3.25 + width/2, -3.25 + width/2], [0, 0], color="k", lw=3)
    ax.plot([-2, -1.5], [-3.25 - width/2, -3.25 - width/2], [0, 0], color="k", lw=3)
    text3d(ax, (-2.5, -2.75, 0),
           r"$V$",
           zdir="y", size=0.5, usetex=True,
           ec="none", fc="k", angle=0.2)
    
# scaling marks
if not AC:
    depth = 0.85
    ax.plot([-1, -1], [0, -depth], [0, 0], color="red", zorder=-1, ls="dashed")
    ax.plot([1, 1], [0, -depth], [0, 0], color="red", zorder=-1, ls="dashed")
    ax.plot([-1, 1], [-depth, -depth], [0, 0], color="red", zorder=-1)
    ax.plot([-1, -1], [(-depth - 0.1), -depth + 0.1], [0, 0], color="red", zorder=-1)
    ax.plot([1, 1], [(-depth - 0.1), -depth + 0.1], [0, 0], color="red", zorder=-1)
    # bb = plt.Rectangle([0.2, -depth - 0.1], 0.65, 0.2, color="white", zorder=10)
    # ax.add_patch(bb)
    # art3d.pathpatch_2d_to_3d(bb, z=0, zdir="z")
    text3d(ax, (0.25, -depth - 0.5, 0),
               r"$2R$",
               zdir="z", size=0.6, usetex=True,
               ec="none", fc="red", angle=0, zorder=100)

# projection lines
ss = 0.1
sb = np.linspace(-ss, ss, 1000)
sbc = "k"
ax.plot([ss, ss], [ss, -ss], color=sbc, ls="dashed")
ax.plot([-ss, ss], [ss, ss], color=sbc, ls="dashed")
ax.plot([-ss, ss], [-ss, -ss], color=sbc, ls="dashed")
ax.plot([-ss, -ss], [-ss, ss], color=sbc, ls="dashed")
rect = plt.Rectangle((-ss, -ss), 2 * ss, 2* ss, fc="k", alpha=0.5)
ax.add_patch(rect)
art3d.pathpatch_2d_to_3d(rect, z=0, zdir="z")
ax.plot([-ss, X1[0, 0]], [-ss, Y1[0, 0]], [0, Z1[0, 0]], color=sbc, zorder=-1, ls="dashed")
ax.plot([-ss, X1[0, -1]], [ss, Y1[0, -1]], [0, Z1[0, -1]], color=sbc, zorder=-1, ls="dashed")
ax.plot([ss, X1[-1, 0]], [-ss, Y1[-1, 0]], [0, Z1[-1, 0]], color=sbc, zorder=100, ls="dashed")
ax.plot([ss, X1[-1, -1]], [ss, Y1[-1, -1]], [0, Z1[-1, -1]], color=sbc, zorder=100, ls="dashed")


ax.set_aspect("equal")
ax.view_init(24, -60)
plt.axis("off")
plt.tight_layout()
if AC:
    plt.savefig("/Users/josephbroz/Desktop/thesis/thesis-images/2dquadrupole_in_3d_ac.pdf")
else:
    plt.savefig("/Users/josephbroz/Desktop/thesis/thesis-images/2dquadrupole_in_3d_dc.pdf")
plt.show()
```
